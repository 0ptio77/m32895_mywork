Question 1 - Prime number Checker

```python
num = int(input("What is your number? ")) #sets num to inputted number

if num < 2:
    print("is not a prime number")    #1 is not a prime number
else:
    for i in range(2, num):           #repeats from 2 to the number searching for a factor
        if num % i == 0:
            print(num, "is not a prime number") #if a factor is found, the loop is broken
            break
    else:
        print(num, "is a prime number!")        #if not, its a prime number
```

Question 2 - fibonaccii sequence generator

```python
def fibonacci(n):                     #defines function 'fibonacci'
    sequence = [0, 1]                 #sets first two values to be 0 and 1
    for _ in range(n - 2):            
        sequence.append(sequence[-1] + sequence[-2]) #sets next value to be the prior two added together
    return sequence[:n]

num_terms = int(input("Enter the number of Fibonacci terms: "))
print(fibonacci(num_terms))        #will repeat the function the required number of times
```

Question 6 - Guess the number game

```python
import random                                  
num = random.randint(1,10)                  #generates random number from 1-10
while True:                                 #infinte loop
    guess = int(input("guess the number"))  #sets guess to a user input
    if guess == num:
        print("Bingo!")                     #breaks loop if the guess is equal to the random number
        break
    elif guess < num:
        print("nope", guess,"is too low")   # if not, checks to see if the guess is higher or lower than the random number and repeats
    else:
        print ("nope",guess,"is too high")

```

Question 8 - formula

```python
import math

x = int(input("X?"))         #asks for user inputs for both X and N
N = int(input("N?"))

def formula_function(x, N):
    total = 0                         #sets starting total to 0
    for i in range(1, N + 1):         #repeats function required number of times
         numerator = x * ((5 ** (i))**2)                           # x*((5^i)^2)
         denominator = i + math.factorial(i + 1)                  # i+(i+1!) 
         term = numerator / denominator 
         total = total + term                                     #adds answer to total
    return total


answer = formula_function(x, N)

print(answer)
```

Question 9 - rotating image

```python
import matplotlib.pyplot as plt
import numpy as np                                     
import PIL
import urllib

cat_url = "https://upload.wikimedia.org/wikipedia/commons/9/95/A_cat_from_lqw.jpg"
image_data = urllib.request.urlopen(cat_url)
image = PIL.Image.open(image_data)                    #Load image from URL

normal_cat = np.array(image)                                     
horizontal_cat = np.fliplr(normal_cat)                 #creates 3 versions of the image normal, horizontally flipped and vertically flipped
vertical_cat = np.flipud(normal_cat)

fig, axes = plt.subplots(1, 3, figsize=(15, 5))        #creates figure with 3 subplots
axes[0].imshow(normal_cat)
axes[1].imshow(horizontal_cat)                         #sets images to corrosponding subplots
axes[2].imshow(vertical_cat)
```

Question 10 - Temperature array

```python
import numpy as np

temperature_data = [22, 24, 19, 21, 23, 25, 26, 27, 24, 23, 20, 22, 21, 19, 18, 22, 23, 24, 
                   26, 28, 30, 27, 25, 26, 24, 22, 21, 20, 19, 23]

temp_array = np.array(temperature_data)        #creates numpy array

min_temp = np.min(temp_array)                  #finds minimun maximum and mean temperatures
max_temp = np.max(temp_array)
avg_temp = np.mean(temp_array)

fahrenheit_array = (temp_array * 1.8) + 32      #coverts array to fahrenheit

temp_above_25 = temp_array[temp_array > 25]     #lists all temperatures above 25 degrees

days_below_20 = np.sum(temp_array < 20)         #lists all temperatures below 20 degrees

sorted_temps = np.sort(temp_array)              #sorts temperatures in acending order

reshaped_temp = temp_array.reshape(5, 6)        #reshapes array to 5x6

print("Minimum temp:", min_temp)
print("Maximum temp:", max_temp)
print("Average temp:", avg_temp)
print("Fahrenheit temp", fahrenheit_array)
print("Days with temperature above 25:", temp_above_25)
print("Number of days with temperature below 20:", days_below_20)
print("Temperatures in order", sorted_temps)
print("Weekly Temperature")
print(reshaped_temp)
```
